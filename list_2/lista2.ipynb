```python
import numpy as np
from scipy.linalg import solve, det, inv
```

## Zadanie 4

```python
A_4 = np.array([
    [0, 0, 2, 1, 2],
    [0, 1, 0, 2, -1],
    [1, 2, 0, -2, 0],
    [0, 0, 0, -1, 1],
    [0, 1, -1, 1, -1]
])

b_4 = np.array([1, 1, -4, -2, -1])
```

```python
def gaussian_elimination(A, b):
    A_mod = A.copy()
    b_mod = b.copy()
    A_mod = A_mod.astype(float)
    b_mod = b_mod.astype(float)
    n = len(b)
    s = 0
    

    for i in range(n):
        max_row = np.argmax(np.abs(A_mod[i:, i])) + i
        if A_mod[max_row, i] == 0:
            raise ValueError("Macierz osobliwa — brak jednoznacznego rozwiązania.")
        
        if max_row != i:
            A_mod[[i, max_row]] = A_mod[[max_row, i]]
            b_mod[[i, max_row]] = b_mod[[max_row, i]]
            s += 1
        
        for j in range(i + 1, n):
            m = A_mod[j, i] / A_mod[i, i]
            A_mod[j, i:] -= m * A_mod[i, i:]
            b_mod[j] -= m * b_mod[i]
    
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        suma = np.dot(A_mod[i, i + 1:], x[i + 1:])
        x[i] = (b_mod[i] - suma) / A_mod[i, i]
    
    return x, A_mod, s
```

```python
x_4, At_4, _ = gaussian_elimination(A_4, b_4)
print(x_4)
```

```python
x_4_scipy = solve(A_4, b_4)
print(x_4_scipy)
```

## Zadanie 8

```python
def norm1(A):
    return np.max(np.sum(abs(A), axis=0))
```

```python
def find_inv_A(A):
    n = len(A)
    I = np.eye(n)
    inv_A = np.zeros((n, n))
    for i in range(len(I)):
        x, _, _ = gaussian_elimination(A, I[i])
        inv_A[:, i] = x
    return inv_A
```

```python
A_8 = np.array([
    [2, -1, 0, 0, 0, 0],
    [-1, 2, -1, 0, 0, 0],
    [0, -1, 2, -1, 0, 0],
    [0, 0, -1, 2, -1, 0],
    [0, 0, 0, -1, 2, -1],
    [0, 0, 0, 0, -1, 5]
])
```

```python
A_8_inv = find_inv_A(A_8)
print(A_8_inv)
```

```python
A_8_inv_scipy = inv(A_8)
print(A_8_inv_scipy)
```

```python
print(A_8 @ A_8_inv)
print(A_8 @ A_8_inv_scipy)
```

```python
print(norm1(A_8))
```
