## Zadanie 1

```python
def z1(x):
    return (6-2*x)/(6+4*x+x**2)

def z2(x):
    return (6-4*x+x**2)/(6+2*x)
```

```python
x = np.linspace(-1, 1, 1000)
plt.plot(x, z1(x), label='z1')
plt.plot(x, z2(x), label='z2')
plt.plot(x, np.exp(-x), linestyle='--', label='exp(-x)')
plt.legend()
plt.grid()
plt.title("Przybli≈ºenia Padego funkcji exp(-x)")
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

x = np.linspace(-5, 5, 1000)
axes[0].plot(x, z1(x), label='z1')
axes[0].plot(x, np.exp(-x), linestyle='--', label='exp(-x)', color='red')
axes[0].legend()
axes[0].grid()
axes[0].set_title("Przybli≈ºenie z1(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")

axes[1].plot(x, z2(x), label='z2')
axes[1].plot(x, np.exp(-x), linestyle='--', label='exp(-x)', color='red')
axes[1].legend()
axes[1].grid()
axes[1].set_title("Przybli≈ºenie z2(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f(x)")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("B≈ÇƒÖd bezwzglƒôdny przybli≈ºe≈Ñ Padego")
x = np.linspace(-5, 5, 1000)

axes[0].plot(x, np.abs(z1(x)- np.exp(-x)), label='z1')
axes[0].set_title("Przybli≈ºenie z1(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].legend()
axes[0].grid()

axes[1].plot(x, np.abs(z2(x) - np.exp(-x)), label='z2')
axes[1].set_title("Przybli≈ºenie z2(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f(x)")
axes[1].legend()
axes[1].grid()

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("B≈ÇƒÖd wzglƒôdny przybli≈ºe≈Ñ Padego")
x = np.linspace(-5, 5, 1000)

axes[0].plot(x, (np.abs(z1(x)- np.exp(-x)))/np.exp(-x), label='z1')
axes[0].set_title("Przybli≈ºenie z1(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].legend()
axes[0].grid()

axes[1].plot(x, (np.abs(z2(x) - np.exp(-x)))/np.exp(-x), label='z2')
axes[1].set_title("Przybli≈ºenie z2(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f(x)")
axes[1].legend()
axes[1].grid()

plt.tight_layout()
```

## Zadanie 4

```python
znak = 0
wykladnik = "01111111"
ulamek = "10110011001100110011001"

bits = f"{znak}{wykladnik}{ulamek}"
bits = bits.ljust(32, '0')

int_bits = int(bits, 2)

approx = struct.unpack('!f', struct.pack('!I', int_bits))[0]

true_value = 1.7

blad_bezwzgl = abs(true_value - approx)
blad_wzgl = blad_bezwzgl / abs(true_value)

print("Reprezentacja bitowa:", bits)
print("Warto≈õƒá przybli≈ºona:", approx)
print("B≈ÇƒÖd bezwzglƒôdny:", blad_bezwzgl)
print("B≈ÇƒÖd wzglƒôdny:", blad_wzgl)
```

## Zadanie 5

```python
def W(x):
    return 6*x**4 + 5*x**3 - 13*x**2 + x + 1
```

```python
x = np.arange(-10, 10, 0.0001)
plt.plot(x, W(x))
plt.xlabel('x')
plt.ylabel('W(x)')
plt.title("W(x) - zwyk≈Çy spos√≥b")
plt.grid()
cProfile.run('W(x)')
```
